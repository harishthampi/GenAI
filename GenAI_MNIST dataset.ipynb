```python
import tensorflow as tf
from tensorflow.keras import layers,models #Keras API included with TensorFlow.
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.mnist.load_data() # This function is used to load the MNIST dataset, one of the most common datasets for training and testing image classification models.
```

```python
  X_train[0] # 0 represents black and 256 represents white
```

```python
plt.imshow(X_train[0],cmap='grey')
```

```python
# scaling down the data from 0 to 255 to 0 to 1
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
X_train[0]
```

```python
# Create a  fully connected neural network  model
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)), # Converts each 28×28 image into a 1 dimensional array.This is the input layer(required to feed images into Dense layers.)
    layers.Dense(128,activation='relu'), # Fully connected hidden layer with 128 neurons( a hyper paramter obtained by trail and error).Activation: ReLU (Rectified Linear Unit) Helps mitigate the vanishing gradient problem also Allows learning non-linear patterns.
    layers.Dense(10, activation='softmax') # Output layer with 10 neurons (one for each digit 0–9).Activation: Softmax Produces a probability distribution over the 10 classes.Sum of outputs = 1.0.
])
```

```python
#compile the model before training:
model.compile(optimizer='adam', # Popular choice, adapts learning rate.
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']) # Track accuracy during training/evaluation.
```

```python
# Training the model
history = model.fit(X_train,y_train,epochs=5)
```

```python
# Testing the model
test_loss , test_accuracy = model.evaluate(X_test,y_test)
print(f'Test accuracy: {test_accuracy*100:.2f}%')
```

```python
predictions = model.predict(X_test)
print(predictions)
print('predictions shape',predictions.shape)
```

```python
# What does model.predict() return?
# Output layer -> layers.Dense(10, activation='softmax') ->outputs 10 probabilities per sample (one for each digit 0–9).
# predictions is a NumPy array with shape (num_samples, 10).MNIST test data:num_samples = 10,000.
```
